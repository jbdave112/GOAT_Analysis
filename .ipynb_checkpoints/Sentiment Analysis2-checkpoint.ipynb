# Cyberpunk 2077 Sentiment Analysis

```python
#Import Necessary Packages
import pandas as pd
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import seaborn as sns
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
import math
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer
from spacy.lang.en import English
```

```python
#Read in Reddit Data
rd_comments = pd.read_csv('rd_comments.csv')
rd_comments = rd_comments.T
rd_comments.columns = ['Comment']
rd_comments['Site'] = 'Reddit'
```

```python
#Read in YouTube Data
yt_comments = pd.read_csv('yt_comments.csv')
yt_comments['Site'] = 'YouTube'
yt_comments = yt_comments[['Site', 'Comment']]
```

```python
#Concat Data
comments_df = pd.concat([rd_comments, yt_comments], sort=True)
comments_df = comments_df[['Comment', 'Site']]
comments_df.head()
```

```python
#Create Vader Sentiment Analyzer
sia = SentimentIntensityAnalyzer()
```

```python
#Function that creates sentiment score
def Sent_analyze(Comment):
        try:
            x = sia.polarity_scores(Comment)['compound']
            return x
        except:
            pass
```

```python
#Create column containing sentiment score
comments_df['Sentiment'] = comments_df['Comment'].apply(Sent_analyze)
```

```python
#Create function to bucket sentiment
def sent_class(x):
    if x > 0.75:
        return 'Very Positive'
    elif 0.75 >= x > 0.25:
        return 'Positive'
    elif 0.25 >= x >= -0.25:
        return 'Neutral'
    elif -0.25 > x >= -0.75:
        return 'Negative'
    elif x < -0.75:
        return 'Very Negative'
    else:
        return 'Other'
```

```python
#Apply function to dataframe
comments_df['sent_class'] = comments_df['Sentiment'].apply(sent_class)
comments_df.head()
```

```python
#Plot Sentiment by Site
fig, ax = plt.subplots(figsize=(8,8))
ax = sns.violinplot(x = 'Site', y = 'Sentiment', data = comments_df)
```

```python
#Plot Count of each class
fig, ax = plt.subplots(figsize=(9,7))
ax = sns.countplot(x='sent_class', data = comments_df, order = ['Very Positive', 'Positive', 'Neutral', 'Negative', 'Very Negative'])
total = float(len(comments_df))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height + 0.5,
            '{:1.0%}'.format(height/total),
            ha="center") 
```

```python
#Examine Negative Comments
pd.set_option('display.max_colwidth', -1)
comments_df[comments_df['sent_class'] == 'Negative'].sort_values(['Sentiment'])['Comment'].head(10)
```

```python
#Function to read in stopwords file
def get_stop_words(stop_file_path):
    with open(stop_file_path, 'r', encoding = 'utf-8') as f:
        stopwords = f.readlines()
        stop_set = set(m.strip() for m in stopwords)
        return frozenset(stop_set)
    
stopwords = get_stop_words('stopwords.txt')
```

```python
#Create word count dataframe; because of the volume of YouTube comments we are going to focus on Reddit
counts_df = comments_df[comments_df['Site'] == 'Reddit']['Comment'].apply(lambda x: pd.value_counts(x.split(" "))).sum(axis=0)
```

```python
#Create wordcounts dataframe
counts_df = counts_df.to_frame()

counts_df = counts_df[counts_df[0] >1]
counts_df = counts_df[counts_df[0].notnull()]
counts_df = counts_df.reset_index()
```

```python
#Create Wordcloud
wordcloud = WordCloud(
    width = 3000,
    height = 2000,
    background_color = 'black',
    stopwords = stopwords).generate(str(counts_df['index']))
fig = plt.figure(
    figsize = (15, 10),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wordcloud, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
